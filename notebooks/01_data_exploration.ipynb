```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# загрузка данных
df = pd.read_csv('../data/raw/creditcard.csv')
df.head()
```

```python
df.shape
df.info()
```

```python
# классы очень несбалансированные
df['Class'].value_counts()
```

```python
# попробую посмотреть распределение Amount
plt.hist(df['Amount'], bins=50)
plt.show()
# хм, не видно ничего из-за выбросов
```

```python
# попробую логарифм
df['Amount_log'] = np.log1p(df['Amount'])
plt.hist(df['Amount_log'], bins=50)
plt.show()
```

```python
# корреляции — попробую посмотреть что там с V-признаками
corr = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr)
plt.show()
# слишком много признаков, не видно нихрена
```

```python
# может посмотрю только на топ корреляций с Class?
corr_with_class = corr['Class'].sort_values(ascending=False)
print(corr_with_class.head(10))
print(corr_with_class.tail(10))
```

```python
# попробую посмотреть временные паттерны
# сгруппирую по часам и посмотрю распределение классов
df['Hour'] = (df['Time'] / 3600).astype(int)
hourly = df.groupby('Hour')['Class'].agg(['sum', 'count', 'mean'])
hourly.plot(figsize=(12, 4))
plt.show()
```
